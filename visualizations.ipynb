```python
import pandas as pd
import matplotlib
import seaborn as sns
```

```python
file = 'NBA_Player_Stats_2.csv'
df = pd.read_csv(file)
print(df.head())
```

```python
df1 = df[df['MVP']  == True]
df1
```

```python
df1['season-year'] = df1['Season'].str[:4]
df1['season-year'] = pd.to_numeric(df1['season-year'])
df1
```

```python
df2 = df1[df1['season-year'] >= 2008]
df2 = df1[df1['season-year'] <= 2017]
df2
```

```python
ax = df2.plot.bar(x = 'Player', y = 'PTS')
```

```python
ax = df2.plot.bar(x = 'Player', y = 'AST')
```
